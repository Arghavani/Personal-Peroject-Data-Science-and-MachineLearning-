```python
#time input(t) 
#sin cos, tan, cot
import numpy as np
import matplotlib.pyplot as plt
t = np.arange(0,5,0.01)
#print(t)
y=np.sin(2*(np.pi)*t)
plt.plot(t,y)
plt.xlabel('time')
plt.ylabel('place')
plt.title('sin')
plt.show()
```
