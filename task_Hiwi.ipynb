## Add colmun zikv_tri1 w.r.t first condition 
## if the patient follow the first condition give me 'True' otherwise give me 'False'
## in this way there is no empty cell in dataframe 

```python
df_copy['zikv_tri1'] =''
df_copy.loc[list(df_copy.loc[df_copy['ga']> 0].index & df_copy.loc[df_copy['ga']<14].index),'zikv_tri1'] = 'True'
df_copy.loc[list(df_copy.loc[df_copy['ga']>= 14].index), 'zikv_tri1']='False'
```

```python
df_copy
```

## Similarly for Zikv_tri2 only the patient who is folllowing the second condition(trimester2) get the value 'True' and others get value 'False'

```python
df_copy['zikv_tri2'] =''
df_copy.loc[list(df_copy.loc[df_copy['ga']> 14].index & df_copy.loc[df_copy['ga']<28].index),'zikv_tri2'] = 'True'
df_copy.loc[list(df_copy.loc[df_copy['ga']<= 14].index), 'zikv_tri2'] = 'False'
df_copy.loc[list(df_copy.loc[df_copy['ga']>= 28].index), 'zikv_tri2'] = 'False'
```

```python
df_copy
```

## Likewise above, third condition for patient is 'True' only if they are in the third condition otherwise is 'False'

```python
df_copy['zikv_tri3'] =''
df_copy.loc[list(df_copy.loc[df_copy['ga']>= 28].index),'zikv_tri3'] = 'True'
df_copy.loc[list(df_copy.loc[df_copy['ga']< 28].index),'zikv_tri3'] = 'False'
```
